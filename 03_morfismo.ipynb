# Isomorfismo

Um mapa entre dois grupos $G$ e $G'$ é dito isomórfico se para todos os pares $g_{j},g_{k}\in G$ e $g'_{j},g'_{k}\in G'$ com correspondência um pra um
\begin{equation}
g_{j} \mapsto g'_{j}\text{ e } g_{k}\mapsto g'_{k},
\end{equation}
temos que a composição é preservada, i.e.,
\begin{equation}
g_{j}*g_{k}\mapsto g'_{j}*g'_{k}.
\end{equation}
Acima e doravante, lê-se $\mapsto$ como "corresponde a" ou "é mapeado em".


## Grupo de simetria de um _triângulo equilátero_, $G_{\triangle}$
Consideremos as transformações de simetria que levam um triângulo equilátero para a mesma orientação espacial inicial, sem considerar a indexação dos vértices feita na figura abaixo.
<img src="fig_03_triangle.png" width="200">
Denotemos $R^{X}_{\theta}$ uma rotação em torno do eixo $X$ por um ângulo $\theta$ (usa a "regra da mão direita"). Os elementos de $G_{\triangle}$ são:
\begin{equation}
G_{\triangle}\equiv\{R^{X}_{0}, R^{O}_{2\pi/3}, R^{O}_{4\pi/3}, R^{A}_{\pi}, R^{B}_{\pi}, R^{C}_{\pi}\}.
\end{equation}
Denotemos por $({}_{1}3_{2})$ a disposição atual do triângulo, ou seja, indexamos os vértices no sentido anti-horário começando com o da esquerda. Então
\begin{align}
& R^{X}_{0}({}_{j}l_{k}) = ({}_{j}l_{k}), \\
& R^{O}_{2\pi/3}({}_{j}l_{k}) = ({}_{l}k_{j}), \\
& R^{O}_{4\pi/3}({}_{j}l_{k}) = ({}_{k}j_{l}), \\
& R^{A}_{\pi}({}_{j}l_{k}) = ({}_{j}k_{l}), \\
& R^{B}_{\pi}({}_{j}l_{k}) = ({}_{l}j_{k}), \\ 
& R^{C}_{\pi}({}_{j}l_{k}) = ({}_{k}l_{j}).
\end{align}
A tabela de composição das transformações de simetria do triângulo é:
$$
\begin{array}{c | c c c c c c}
& R^{X}_{0} & R^{O}_{2\pi/3} & R^{O}_{4\pi/3} & R^{A}_{\pi} & R^{B}_{\pi} & R^{C}_{\pi}  \\
\hline
R^{X}_{0} & R^{X}_{0} & R^{O}_{2\pi/3} & R^{O}_{4\pi/3} & R^{A}_{\pi} & R^{B}_{\pi} & R^{C}_{\pi} \\
R^{O}_{2\pi/3} & R^{O}_{2\pi/3} & R^{O}_{4\pi/3} & R^{X}_{0} & R^{C}_{\pi} & R^{A}_{\pi} & R^{B}_{\pi} \\
R^{O}_{4\pi/3} & R^{O}_{4\pi/3} & R^{X}_{0} & R^{O}_{2\pi/3} & R^{B}_{\pi} & R^{C}_{\pi} & R^{A}_{\pi} \\
R^{A}_{\pi} & R^{A}_{\pi} & R^{B}_{\pi} & R^{C}_{\pi} & R^{X}_{0} & R^{O}_{2\pi/3} & R^{O}_{4\pi/3} \\
R^{B}_{\pi} & R^{B}_{\pi} & R^{C}_{\pi} & R^{A}_{\pi} & R^{O}_{4\pi/3} & R^{X}_{0} & R^{O}_{2\pi/3} \\
R^{C}_{\pi} & R^{C}_{\pi} & R^{A}_{\pi} & R^{B}_{\pi} & R^{O}_{2\pi/3} & R^{O}_{4\pi/3} & R^{X}_{0}
\end{array}.
$$
A verificação dessa tabela é mostrada a seguir.

A 1ª linha e a 1ª coluna não precisam ser verificadas pois $R_{0}^{x}=e$, e assim $X*R_{0}^{x}=R_{0}^{x}*X=X$ para todo $X\in G_{\triangle}$. 

Para a 2ª linha fazemos
\begin{align}
& R_{2\pi/3}^{O}*R_{2\pi/3}^{O}({}_{j}l_{k}) = R_{2\pi/3}^{O}({}_{l}k_{j}) = ({}_{k}j_{l}) \equiv R_{4\pi/3}^{O}({}_{j}l_{k}) \ \therefore \ R_{2\pi/3}^{O}*R_{2\pi/3}^{O}=R_{4\pi/3}^{O}, \\
& R_{2\pi/3}^{O}*R_{4\pi/3}^{O}({}_{j}l_{k}) = R_{2\pi/3}^{O}({}_{k}j_{l}) = ({}_{j}l_{k}) \equiv R_{0}^{X}({}_{j}l_{k}) \ \therefore \ R_{2\pi/3}^{O}*R_{4\pi/3}^{O}=R_{0}^{X}, \\
& R_{2\pi/3}^{O}*R_{\pi}^{A}({}_{j}l_{k}) = R_{2\pi/3}^{O}({}_{j}k_{l}) = ({}_{k}l_{j}) \equiv R_{\pi}^{C}({}_{j}l_{k}) \ \therefore \ R_{2\pi/3}^{O}*R_{\pi}^{A}=R_{\pi}^{C}, \\
& R_{2\pi/3}^{O}*R_{\pi}^{B}({}_{j}l_{k}) = R_{2\pi/3}^{O}({}_{l}j_{k}) = ({}_{j}k_{l}) \equiv R_{\pi}^{A}({}_{j}l_{k}) \ \therefore \ R_{2\pi/3}^{O}*R_{\pi}^{B}=R_{\pi}^{A}.
\end{align}
Pelo teorema do rearranjamento, o último elemento da 2ª linha deve ser $R_{\pi}^{B}.$

Para a 2ª linha, começamos notando que $R_{4\pi/3}^{O}*R_{2\pi/3}^{O}=R_{0}^{X}$, pois a composição das rotações de $120^{\circ}$ e $240^{\circ}$ em torno de $O$ corresponde a uma rotação de $360^{\circ},$ que traz o triângulo para sua disposição inicial. Também teremos que $R_{4\pi/3}^{O}*R_{4\pi/3}^{O}=R_{2\pi/3}^{O}$, pois a composição das rotações de $240^{\circ}$ e $240^{\circ}$ em torno de $O$ corresponde à composição de rotações por $360^{\circ}$ e $120^{\circ}.$
Para as outras composições
\begin{align}
& R_{4\pi/3}^{O}*R_{\pi}^{A}({}_{j}l_{k}) = R_{4\pi/3}^{O}({}_{j}k_{l}) = ({}_{l}j_{k}) \equiv R_{\pi}^{B}({}_{j}l_{k}) \ \therefore \ R_{4\pi/3}^{O}*R_{\pi}^{A}=R_{\pi}^{B}, \\
& R_{4\pi/3}^{O}*R_{\pi}^{B}({}_{j}l_{k}) = R_{4\pi/3}^{O}({}_{l}j_{k}) = ({}_{k}l_{j}) \equiv R_{\pi}^{C}({}_{j}l_{k}) \ \therefore \ R_{4\pi/3}^{O}*R_{\pi}^{B}=R_{\pi}^{C}, \\
\end{align}
Pelo teorema do rearranjamento, o último elemento da 3ª linha deve ser $R_{\pi}^{A}.$

Para a 4ª linha teremos
\begin{align}
& R_{\pi}^{A}*R_{2\pi/3}^{O}({}_{j}l_{k}) = R_{\pi}^{A}({}_{l}k_{j}) = ({}_{l}j_{k}) \equiv R_{\pi}^{B}({}_{j}l_{k}) \ \therefore \ R_{\pi}^{A}*R_{2\pi/3}^{O}=R_{\pi}^{B}, \\
& R_{\pi}^{A}*R_{4\pi/3}^{O}({}_{j}l_{k}) = R_{\pi}^{A}({}_{k}j_{l}) = ({}_{k}l_{j}) \equiv R_{\pi}^{C}({}_{j}l_{k}) \ \therefore \ R_{\pi}^{A}*R_{4\pi/3}^{O}=R_{\pi}^{C}, \\

\end{align}
Pelo teorema do rearranjamento, o último elemento da 4ª linha deve ser $R_{4\pi/3}^{O}.$

Para a 5ª linha teremos
\begin{align}
& R_{\pi}^{B}*R_{2\pi/3}^{O}({}_{j}l_{k}) = R_{\pi}^{B}({}_{l}k_{j}) = ({}_{k}l_{j}) \equiv R_{\pi}^{C}({}_{j}l_{k}) \ \therefore \ R_{\pi}^{B}*R_{2\pi/3}^{O}=R_{\pi}^{C}, \\

\end{align}
Pelo teorema do rearranjamento, o último elemento da 4ª linha deve ser $R_{2\pi/3}^{O}.$

A 6ª linha da tabela de composição pode ser obtida aplicando-se o teorema do rearranjamento a cada coluna.

## Grupo de permutação de três índices, $S_{3}$
A maneira como escrevemos as tranformações de simetria e disposições do triângulo acima indicam uma correspondência isomórfica entre esse grupo e o grupo de permutações de três índices, denotado por $S_{3}.$ Nesse caso os elementos são denotados por
\begin{equation}
P_{pqr}.
\end{equation}
Quando este $P_{pqr}$ atua em uma lista $(j,k,l)$ de números diferentes retirados do conjunto $\{1,2,3\}$, este coloca o símbolo que está na posição $p$ para a posição $1$, o símbolo da posição $q$ vai para a posição $2$ e o símbolo da posição $r$ vai para a posição $3$. \begin{align}

& P_{1,3,2}(j,k,l) = (j,l,k), \\
& P_{2,1,3}(j,k,l) = (k,j,l), \\
& P_{2,3,1}(j,k,l) = (k,l,j), \\
& P_{3,1,2}(j,k,l) = (l,j,k), \\
& P_{3,2,1}(j,k,l) = (l,k,j).
\end{align}

A tabela de composição para o grupo $S_{3}$ é:
$$
\begin{array}{c | c c c c c c}

## Grupos finitos e o grupo de permutações

### Teorema de Cayley
Todo grupo de ordem finita é isomorfo ao grupo de permutações de $n$ índices, $S_{n}$.

#### Prova
Considere um grupo finito $G=\{g_{1},g_{2},\cdots,g_{n}\}$. Pelo teorema do rearranjamento sabemos que para qualquer $g_{j}\in G$ vem que
\begin{align}
& g_{j}*G=\{g_{j}*g_{1},g_{j}*g_{2},\cdots,g_{j}*g_{n}\} \\
& \equiv \{g_{j_{1}},g_{j_{2}},\cdots,g_{j_{n}}\} \equiv G.
\end{align}
Ou seja, a composição com um elemento de $G$ rearranja a sequência de índices dos elementos de $G$. Baseados nessas observações, definimos a seguinte correspondência isomórfica entre $G$ e $S_{n}$:
\begin{equation}
g_{j} \mapsto P_{g_{j}}\equiv P_{g_{j}*g_{1},g_{j}*g_{2},\cdots,g_{j}*g_{n}} \equiv P_{j_{1},j_{2},\cdots,j_{n}},
\end{equation}
onde $$j_{k} \text{ é obtido de } g_{j}*g_{k}=g_{j_{k}}.$$ 

Como $e*g_{k}=g_{k}\Rightarrow j_{k}=k\ \forall k$, temos que
\begin{equation}
e\mapsto P_{1,2,\cdots,n}.
\end{equation}

Com relação à composição, pela associatividade em $G$, teremos
\begin{align}
P_{g_{j}*g_{k}}(1,\cdots,n) & \equiv P_{(g_{j}*g_{k})*g_{1},\cdots,(g_{j}*g_{k})*g_{n}} = P_{g_{j}*(g_{k}*g_{1}),\cdots,g_{j}*(g_{k}*g_{n})} \\ 
& = P_{g_{j}*g_{k_{1}},\cdots,g_{j}*g_{k_{n}}} = P_{g_{j}}(k_{1},\cdots,k_{n}) = P_{g_{j}}(P_{g_{k}}(1,\cdots,n)) \\
& =(P_{g_{j}}*P_{g_{k}})(1,\cdots,n).
\end{align}
Portanto $$g_{j}*g_{k} \mapsto P_{g_{j}*g_{k}}.$$

Associatividade segue de
\begin{align}
P_{g_{j}}*(P_{g_{k}}*P_{g_{l}}) &= P_{g_{j}}*P_{g_{k}*g_{l}} = P_{g_{j}*(g_{k}*g_{l})} = P_{(g_{j}*g_{k})*g_{l}} \\
& = P_{g_{j}*g_{k}}*P_{g_{l}} = (P_{g_{j}}*P_{g_{k}})*P_{g_{l}}.
\end{align}

Além disso, $g_{j}^{-1}\mapsto P_{g_{j}^{-1}}$ e
\begin{align}
P_{g_{j}^{-1}}*P_{g_{j}}(1,\cdots,n) & = P_{g_{j}^{-1}*g_{j}}(1,\cdots,n) = P_{(g_{j}^{-1}*g_{j})*g_{1},\cdots,(g_{j}^{-1}*g_{j})*g_{n}} = P_{e*g_{1},\cdots,e*g_{n}} \\ 
& = P_{1,2,\cdots,n}(1,\cdots,n).
\end{align}

Como o mapa definido acima é claramente um pra um, a correspondência é vista como sendo isomórfica. $\hspace{2cm}\square$

# Homomorfismo
Um mapa entre dois grupos $G$ e $G'$ é dito homomórfico se para $\{g^{(j)}_{m}\}_{m=1}^{N_{j}},\{g^{(k)}_{n}\}_{n=1}^{N_{k}}\in G$ e $g'_{j},g'_{k}\in G'$ com correspondência vários pra um
\begin{equation}
g^{(j)}_{m} \mapsto g'_{j}\text{ e } g^{(k)}_{n}\mapsto g'_{k},
\end{equation}
temos que a composição é preservada, i.e.,
\begin{equation}
g^{(j)}_{m}*g^{(k)}_{n}\mapsto g'_{j}*g'_{k} \text{ }\forall m,n.
\end{equation}

##### Exemplo
Os números complexos não nulos, $G=\mathbb{C}_{\ne 0}$, e os números reais não nulos, $G'=\mathbb{R}_{\ne 0}$, formam grupos sob a operação de multiplicação. Podemos definir um mapa $\infty$ pra um entre esses dois grupos como segue. Para $z\in\mathbb{C}_{\ne 0}$ definimos
\begin{equation}
z\mapsto |z|=\sqrt{z^{*}z}\in\mathbb{R}_{\ne 0}.
\end{equation}
Usando as propriedades da multiplicação de números complexos vemos prontamente que esse mapa preserva composição, i.e., para $w\in\mathbb{C}_{\ne 0}$ teremos
\begin{align}
& z*w\mapsto |z|*|w|=\sqrt{z^{*}z}\sqrt{w^{*}w}=\sqrt{z^{*}zw^{*}w} \\
& = \sqrt{(zw)^{*}(zw)} = |z*w|.
\end{align}

## Algumas consequências da correspondência homomórfica
* O elemento identidade de $G$ é mapeado no elemento identidade de $G'$:
\begin{align}
&g_{m}^{(j)}*e=g_{m}^{(j)}\forall m,j \Rightarrow (g_{m}^{(j)}*e)'=(g_{m}^{(j)})'*(e)'=g'_{j}*(e)'=(g_{m}^{(j)})'=g'_{j}, \\ 
& g'_{j}=g'_{j}*e'\forall j \Rightarrow e\mapsto e'.
\end{align}

* Se $g_{m}^{(j)}\mapsto g'_{j}$ então $(g_{m}^{(j)})^{-1}\mapsto (g'_{j})^{-1}$:
\begin{align}
& g_{m}^{(j)}*(g_{m}^{(j)})^{-1}=e \\
& \Rightarrow (e)'=(g_{m}^{(j)}*(g_{m}^{(j)})^{-1})'=(g_{m}^{(j)})'*((g_{m}^{(j)})^{-1})' \\
& \Rightarrow g'_{j}*((g_{m}^{(j)})^{-1})'=e'\Rightarrow (g'_{j})^{-1}*g'_{j}*((g_{m}^{(j)})^{-1})'=(g'_{j})^{-1}*e' \\
&\Rightarrow ((g_{m}^{(j)})^{-1})'=(g'_{j})^{-1}.
\end{align}

* O conjunto de elementos de $G$ mapeados em $e'$, que denotamos por $H_{e'}$, é um subgrupo invariante de $G$:<br>
Usando a notação $g'_{0}=e'$ teremos composição
\begin{equation}
(g_{m}^{(0)}*g_{n}^{(0)})' = g'_{0}*g'_{0} = e'*e'=e'.
\end{equation}
Associatividade vem daquela de $G$. <br>
Ademais $$e\mapsto e' \text{ e } (g_{m}^{(0)})^{-1}\mapsto(g'_{(0)})^{-1}=e'.$$ Portanto $H_{e'}$ é um subgrupo de $G$. <br>
Verificamos invariância por
\begin{align}
& (g_{n}^{(k)}*g_{n}^{(0)}*(g_{n}^{(k)})^{-1})'=(g_{n}^{(k)})'*(g_{n}^{(0)})'*((g_{n}^{(k)})^{-1})' = g'_{k}*e'*(g'_{k})^{-1} \\
& = g'_{k}*(g'_{k})^{-1}=e'\text{, }\forall n,k.
\end{align}

* $G'$ é isomórfico ao grupo fator $G/H_{e'}=\{g_{1}*H_{e'},\cdots,g_{|G|}*H_{e'}\}$:<br>
Aqui a verificação é feita notando-se que $\forall n,k$
\begin{equation}
(g_{n}^{(k)}*H_{e'})'=(g_{n}^{(k)})'*(H_{e'})'=g'_{k}*\{e'\}=g'_{k}.
\end{equation}

##### Exemplo
Podemos ilustrar essas consequências usando o exemplo $G\equiv\mathbb{C}_{\ne0}$ e $G'\equiv\mathbb{R}_{\ne0}$ com correspondência $z\mapsto|z|$: 
* Como $e=1$ e $e'=1$ temos que $e\mapsto e'$.
* Para o inverso, temos que $z^{-1}=z^{*}/|z|^{2}\mapsto|z^{*}/|z|^{2}|=1/|z|$, que é o elemento inverso de $|z|$.
* Os elementos de $\mathbb{C}_{\ne0}$ mapeados em $e'=1$ tem a forma $z=|z|e^{i\phi}=e^{i\phi}$, e formam um grupo, $H_{e'}$, pois $e^{i\phi}e^{i\phi'}=e^{i\phi''}$. Este é um subgrupo invariante pois
\begin{equation}
z\{e^{i\phi}\}z^{-1}=\{ze^{i\phi}z^{-1}\}=\{e^{i\phi}zz^{-1}\}=\{e^{i\phi}e\}=\{e^{i\phi}\}=H_{e'}\forall z\in\mathbb{C}_{\ne0}.
\end{equation}
* Sejam $\{z_{|z|}\}$ os números complexos com mesmo módulo $|z|$. O grupo fator de $H_{e'}$ em $\mathbb{C}_{\ne0}$ é 
\begin{align}
\mathbb{C}_{\ne0}/H_{e'}& = \{z_{|z|}H_{e'}\} = \{z_{|z|}\{e^{i\phi}\}\} = \{\{z_{|z|}e^{i\phi}\}\} = \{\{|z_{|z|}|e^{i\phi'}e^{i\phi}\}\} \\
& = \{\{|z_{|z|}|e^{i\phi''}\}\}.
\end{align}
Cada elemento do grupo fator é mapeado isomorficamente em um elemento de $\mathbb{R}_{\ne0}$, $|z_{|z|}|$.
