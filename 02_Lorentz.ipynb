# O grupo de Lorentz

## Mecânica Newtoniana (MN) e transformações de Galilei (TG)
### Procedimento geral da mecânica:
* A partir do movimento, de medidas de posição e de tempo, podemos inferir as forças:
\begin{equation}
(\vec{r},t)\rightarrow\vec{v}(t)=\frac{d\vec{r}(t)}{dt}\rightarrow\vec{a}(t)=\frac{d\vec{v}(t)}{dt}\rightarrow\vec{F}(t)=m\vec{a}(t).
\end{equation}
* A partir das forças, e condições iniciais, podemos prever o movimento:
\begin{equation}
\vec{a}(t)=\frac{\vec{F}(t)}{m}\rightarrow \vec{v}(t)=\vec{v}(t_{0})+\int_{t_{0}}^{t}\vec{a}(t')dt'\rightarrow\vec{r}(t)=\vec{r}(t_{0})+\int_{t_{0}}^{t}\vec{v}(t')dt'.
\end{equation}
* Para verificação das previsões, precisamos medir $(\vec{r},t).$

### Descrição do movimento por observadores em movimento relativo
Consideremos dois observadores $O$ e $O'$ em movimento relativo com velocidade constante $\vec{V}$, como ilustrado na figura abaixo.

<img src="fig_02_ref_frame.png" width="400">

__Postulado do tempo absoluto na MN__: O tempo medido por todos os relógios é o mesmo, independente de sua velocidade. Aqui $$t=t'.$$ 

Pela soma vetorial obteremos as chamadas _transformações de Galilei_, que relacionam as coordenadas espaço-temporais de uma partícula em diferentes referenciais inerciais (RIs):
\begin{align}
& \vec{r}=\vec{r}'+\vec{V}t, \\
& \therefore \text{ }\vec{r}'=\vec{r}-\vec{V}t.
\end{align}

_OBS:_ Assumimos _geometria Euclidiana_ em todas as discussões neste tópico.

Não é difícil verificar que a 2ª lei de Newton é invariante por TG:
\begin{align}
\vec{F}' & = m\vec{a}' = m\frac{d^{2}\vec{r}'}{dt'^{2}} \\ 
& = m\frac{d^{2}(\vec{r}-\vec{V}t)}{dt^{2}} = m\frac{d(\vec{v}-\vec{V})}{dt} = m(\vec{a}-\vec{0}) \\
& =m\vec{a}= \vec{F}.
\end{align}
Ou seja, O e O' obtêm as mesmas forças. 

## Mecânica Einsteiniana (ME) e transformações de Lorentz (TL)

Existiam duas motivações fortes para o desenvolvimento de uma mecânica melhor que a Newtoniana:
* As eqs. de Maxwell não são invariantes por TG. 
* O experimento de Michelson e Morley [https://www.nobelprize.org/prizes/physics/1907/michelson/lecture/] indicou que a velocidade da luz no vácuo é a mesma para todos os referenciais.

A seguir apresentamos as suposições básicas usadas para construir essa nova mecânica.

### Postulados da teoria da _relatividade restrita_:
__P1.__ Referenciais inerciais devem ser equivalentes no que concerne as leis físicas.<br>
__P2.__ A velocidade da luz no vácuo é a mesma, $c$, quando medida em todos os refernciais inerciais.

### Tranformações de Lorentz "1D"
Esses postulados nos fornecerão relações distintas das TG para as coordenadas de uma partícula em diferentes RIs. Aqui assumiremos movimento relativo entre os RIs em uma única direção, $x$. Nesse caso as TG ficam 
\begin{align}
& x=x'+V_{x}t'=x'+Vt\text{ }\therefore \text{ }x'=x-Vt, \\
& y=y'+V_{y}t'=y'+0t\text{, }\therefore \text{ }y'=y, \\
& z=z'+V_{z}t'=z'+0t\text{, }\therefore \text{ }z'=z.
\end{align}

Uma possibilidade para obtermos as tranformações de Lorentz é usar uma _correção relativística_ global $\gamma$ para as TG, i.e., teríamos $y=y'$, $z=z'$ e
\begin{equation}
x'=\gamma(x-Vt).
\end{equation}
Por P1 teremos
\begin{equation}
x=\gamma(x'+Vt').
\end{equation}
_OBS._ Aqui não assumimos que o tempo é absoluto, como era o caso na mecânica Newtoniana. <br>
Manipulando essas equações,
\begin{equation}
x=\gamma(x'+Vt')=\gamma(\gamma(x-Vt)+Vt')=\gamma^{2}(x-Vt)+\gamma Vt'.
\end{equation}
Com isso obtemos a seguinte relação entre o tempo medido pelo observador $O'$ com aquele medido por $O$:
\begin{equation}
t'=\gamma\left(t+\frac{1-\gamma^{2}}{\gamma^{2}V}x\right).
\end{equation}
De forma análoga, pode-se verificar que (em conformidade com P1):
\begin{equation}
t=\gamma\left(t'-\frac{1-\gamma^{2}}{\gamma^{2}V}x'\right).
\end{equation}



#### Obtendo $\gamma$
O que ainda nos falta é obter a correção relativística $\gamma$. Pra isso podemos utilizar o seguinte experimento de pensamento. Quando $O$ e $O'$ estão na mesma posição, um pulso esférico de luz é emitido nessa posição. Por P2, as distâncias da frente de onda medidas em O e O' são, respectivamente, dadas por
\begin{align}
& R^{2}=x^{2}+y^{2}+z^{2}=c^{2}t^{2}, \\
& R'^{2}=x'^{2}+y'^{2}+z'^{2}=c^{2}t'^{2}.
\end{align}
Agora, substituímos as coordenadas de um referencial em relação às do outro, manipulamos a equação obtida e exigimos que ela seja equivalente à outra equação. Para $R'^{2}$:
\begin{align}
& R'^{2}=(\gamma(x-Vt))^{2}+y^{2}+z^{2}=c^{2}((\gamma(t-(1-\gamma^{2})x/\gamma^{2}V))^{2}) \\
& \therefore \gamma^{2}x^{2} +\gamma^{2}V^{2}t^{2}-\gamma^{2}2xVt+y^{2}+z^{2}=c^{2}\gamma^{2}t^{2}+c^{2}\gamma^{2}(1-\gamma^{2})^{2}x^{2}/\gamma^{4}V^{2}-c^{2}\gamma^{2}2t(1-\gamma^{2})x/\gamma^{2}V \\
& \therefore Ax^{2} + y^{2}+z^{2}=\gamma^{2}(1-V^{2}/c^{2})c^{2}t^{2} + Bxt := R^{2}. \hspace{1cm}(*)
\end{align}

Pela equivalência com a eq. para $R^{2}$, devemos ter $A=1$, $B=0$ e 
\begin{equation}
\gamma^{2}(1-V^{2}/c^{2})=1\Rightarrow \gamma=\frac{1}{\sqrt{1-V^{2}/c^{2}}}=\gamma(V).
\end{equation}

Como
\begin{equation}
\frac{1-\gamma^{2}}{\gamma^{2}} = \frac{1-\frac{1}{1-V^{2}/c^{2}}}{\frac{1}{1-V^{2}/c^{2}}} = \frac{\frac{1-V^{2}/c^{2}-1}{1-V^{2}/c^{2}}}{\frac{1}{1-V^{2}/c^{2}}} = -\frac{V^{2}}{c^{2}},
\end{equation}
podemos escrever
\begin{equation}
t'=\gamma\left(t-\frac{V}{c^{2}}x\right)\text{ e } t=\gamma\left(t'+\frac{V}{c^{2}}x'\right).
\end{equation}

### Composição relativística de velocidades
Na mecânica Newtoniana, se uma partícula possui velocidade $\vec{u}$($\vec{u}'$) em relação a um referencial inercial $O$($O'$) e a velocidade relativa dos referenciais é $\vec{V}=(V,0,0)$, então
\begin{align}
\vec{u}' &= \left(u_{x}',u_{y}',u_{z}'\right)  = \left(\frac{d x'}{dt'},\frac{d y'}{dt'},\frac{d z'}{dt'}\right) = \left(\frac{d(x-Vt)}{dt},\frac{d y}{dt},\frac{d z}{dt}\right) \\
&= \left(u_{x}-V,u_{y},u_{z}\right).
\end{align}
Vejamos como fica essa relação no caso relativístico. Considere a figura acima com movimento relativo dos referenciais somente na direção $x$. A velocidade da partícula em relação a $O$ é a taxa de variação da posição medida em $O$ com o tempo medido em $O$:
\begin{equation}
\vec{u}(t)=\frac{d\vec{r}(t)}{dt}=\left(\frac{dx(t)}{dt},\frac{dy(t)}{dt},\frac{dz(t)}{dt}\right)=(u_{x}(t),u_{y}(t),u_{z}(t)).
\end{equation}
Em relação a $O'$:
\begin{equation}
\vec{u}'(t')=\frac{d\vec{r}'(t')}{dt'}=\left(\frac{dx'(t')}{dt'},\frac{dy'(t')}{dt'},\frac{dz'(t')}{dt'}\right)=(u_{x}'(t'),u_{y}'(t'),u_{z}'(t')).
\end{equation}
De $x'=\gamma(x-Vt)$, $y'=y$, $z'=z$ e $t'=\gamma(t-(V/c^{2})x)$, usando 
\begin{align}
dx' & = \frac{\partial x'}{\partial x}dx+\frac{\partial x'}{\partial y}dy+\frac{\partial x'}{\partial z}dz+\frac{\partial x'}{\partial t}dt = \gamma dx-\gamma Vdt, \\
dy' & = \frac{\partial y'}{\partial x}dx+\frac{\partial y'}{\partial y}dy+\frac{\partial y'}{\partial z}dz+\frac{\partial y'}{\partial t}dt  = dy, \\
dz' & = \frac{\partial z'}{\partial x}dx+\frac{\partial z'}{\partial y}dy+\frac{\partial z'}{\partial z}dz+\frac{\partial z'}{\partial t}dt  = dz, \\
dt' & = \frac{\partial t'}{\partial x}dx+\frac{\partial t'}{\partial y}dy+\frac{\partial t'}{\partial z}dz+\frac{\partial t'}{\partial t}dt  = -(\gamma V/c^{2})dx+\gamma dt,
\end{align}
obtemos
\begin{align}
& u_{x}'=\frac{dx'(t')}{dt'}=\frac{\gamma(dx-Vdt)}{\gamma(dt-(V/c^{2})dx)} = \frac{u_{x}-V}{1-u_{x}V/c^{2}}, \\
& u_{y}'=\frac{dy'(t')}{dt'}=\frac{dy}{\gamma(dt-(V/c^{2})dx)} = \frac{u_{y}}{\gamma(1-u_{x}V/c^{2})}, \\
& u_{z}'=\frac{dz'(t')}{dt'}=\frac{dz}{\gamma(dt-(V/c^{2})dx)} = \frac{u_{z}}{\gamma(1-u_{x}V/c^{2})}.
\end{align}

## Grupo de Lorentz
Para obter os elementos do grupo de Lorentz, primeiramente representaremos como vetores coluna (quadrivetores) as coordenadas espaço-temporais de uma partícula nos referenciais $O$ e $O'$, respectivamente:
\begin{equation}
\rho=\begin{bmatrix} x \\ y \\ z \\ ct \end{bmatrix} \text{, } \rho'=\begin{bmatrix}x' \\ y' \\ z' \\ ct' \end{bmatrix}.
\end{equation}
Assim, teremos que
\begin{align} 
\rho' &= \begin{bmatrix}x' \\ y' \\ z' \\ ct' \end{bmatrix} = \begin{bmatrix} \gamma(V)(x-Vt) \\ y \\ z \\ c\gamma(V)(t-Vx/c^{2}) \end{bmatrix} 
= \begin{bmatrix} \gamma(V) & 0 & 0 & -\gamma(V)V/c \\ 0&1&0&0 \\ 0&0&1&0 \\ -\gamma(V)V/c & 0 & 0 & \gamma(V) \end{bmatrix}\begin{bmatrix} x \\ y \\ z \\ ct \end{bmatrix} \\
& =: \Lambda(V)\rho,
\end{align} 
onde definimos
\begin{equation}
\Lambda(V)=\begin{bmatrix} \gamma(V) & 0 & 0 & -\gamma(V)V/c \\ 0&1&0&0 \\ 0&0&1&0 \\ -\gamma(V)V/c&0&0&\gamma(V) \end{bmatrix},
\end{equation}
que serão os _elementos do grupo de Lorentz_, cuja operação de composição é o _produto matricial_. Isso porque se um referencial $O''$ se movimenta com velocidade constate $V'$ em relação a $O'$ (com movimento relativo somente na direção $x$), então:
\begin{equation}
\rho''=\Lambda(V')\rho'=\Lambda(V')\Lambda(V)\rho.
\end{equation}
Ademais, se $V''$ é a velocidade relativa entre $O''$ e $O$ (com movimento relativo somente na direção $x$), verificaremos que neste caso
\begin{equation}
\rho''=\Lambda(V'')\rho.
\end{equation}

### Verificação da estrutura de grupo
É fácil ver que o elemento identidade é 
$$\Lambda(0)=\mathbb{I}_{4}=e.$$ 

Usando o código abaixo, podemos verificar que o elemento inverso de $\Lambda(V)$ é $\Lambda(-V)$, i.e., $$\Lambda(V)\Lambda(-V)=\Lambda(-V)\Lambda(V)=\Lambda(0) \therefore (\Lambda(V))^{-1}=\Lambda(-V).$$ Associatividade vem da associatividade do produto matricial. 

Quanto à composição, vem que
\begin{equation}
\Lambda(V)\Lambda(V')= \begin{bmatrix} \gamma(V)\gamma(V')(1+VV'/c^{2})&0&0&-\gamma(V)\gamma(V')(V+V')/c \\ 0&1&0&0 \\ 0&0&1&0 \\ -\gamma(V)\gamma(V')(V+V')/c&0&0&\gamma(V)\gamma(V')(1+VV'/c^{2}) \end{bmatrix}.
\end{equation}
Precisamos ainda verificar se $\Lambda(V)\Lambda(V')=\Lambda(V'')$, com $V''$ sendo a velocidade relativa entre $O''$ e $O.$

```python
L, g, V, c, v = symbols("Lambda gamma V c v") # usa V'=v
def g(V): # correção relativística
    return 1/sqrt(1-V**2/c**2)
def L(V): # elementos do grupo de Lorentz
    return Matrix([[g(V),0,0,-g(V)*V/c],[0,1,0,0],[0,0,1,0],[-g(V)*V/c,0,0,g(V)]])
#L(V), L(0), simplify(L(V)*L(-V)), simplify(L(-V)*L(V))# identidade e inverso
L(V)*L(v) # composição
```

Aplicando a composição relativística de velocidades, descrita acima, para os três referenciais com movimentos relativos como na figura
<img src="ref_frame_.png" width="500">
teremos que a velocidade de $O''$ (a partícula nesse caso) em relação a $O$ será obtida como segue ($u_{y}=u_{z}=0\therefore u_{y}'=u_{z}'=0$):
\begin{align}
& u_{x}'= \frac{u_{x}-V}{1-u_{x}V/c^{2}} \therefore V'= \frac{V''-V}{1-V''V/c^{2}} \therefore V'(1-V''V/c^{2})= V''-V  \\
& \therefore V''=\frac{V+V'}{1+VV'/c^{2}}.
\end{align}

Vamos verificar que $\Lambda(V'')=\Lambda(V)\Lambda(V')$. Para isso devemos ter que
\begin{align}
& \gamma(V'')=\gamma(V)\gamma(V')(1+VV'/c^{2}), \hspace{3.3cm}(\star)\\
& -\gamma(V'')V''/c = -\gamma(V)\gamma(V')(V+V')/c. \hspace{2cm}(\star\star)
\end{align}

Vamos assumir que a Eq. ($\star$) está correta e obter dela a relação certa para $V''$. Começamos invertendo ($\star$) e a elevando ao quadrado:
\begin{align}
& \frac{(1+VV'/c^{2})^{2}}{\gamma(V'')^{2}} = \frac{1}{\gamma(V)^{2}\gamma(V')^{2}} \\
& \Rightarrow (1+VV'/c^{2})^{2}(1-V''^{2}/c^{2}) = (1-V^{2}/c^{2})(1-V'^{2}/c^{2}) \\
& \Rightarrow 1+2VV'/c^{2}+V^{2}V'^{2}/c^{4}-(V''^{2}/c^{2})(1+VV'/c^{2})^{2} = 1-V'^{2}/c^{2}-V^{2}/c^{2}+V^{2}V'^{2}/c^{4} \\
& \Rightarrow V'^{2}+V^{2}+2VV' = (V+V')^{2} = V''^{2}(1+VV'/c^{2})^{2} \\
& \Rightarrow V''=(V+V')/(1+VV'/c^{2}),
\end{align}
que é a expressão correta. Pode-se verificar que a igualdade na Eq. $(\star\star)$ também é verdadeira.
Com isso mostramos fechamento para o grupo de Lorentz.

# Grupo de Poincaré
## Transformações de Lorentz "3D"
Considere a figura abaixo:

<img src="lorentz3D.png" width="400">

Vamos decompor os vetores posição em termos de suas componentes paralelas e ortogonais à velocidade relativa $\vec{V}=(V_{x},V_{y},V_{z})$ entre os referenciais inerciais $O$ e $O'$:
\begin{equation}
\vec{r}=\vec{r}_{\perp}+\vec{r}_{\parallel} \text{ e } \vec{r'}=\vec{r'}_{\perp}+\vec{r'}_{\parallel}.
\end{equation}
Como no caso 1D, aqui as componentes perpendiculares a $\vec{V}$ não são afetadas pelo movimento relativo. Para as componentes paralelas a $\vec{V}$ teremos
\begin{align}
& \vec{r'}_{\parallel}=\gamma(\vec{r}_{\parallel}-\vec{V}t), \\
& \vec{r}_{\parallel}=\gamma(\vec{r'}_{\parallel}+\vec{V}t'). \ \text{(por P1)}
\end{align}
Substituindo uma equação na outra obteremos
\begin{align}
& \vec{r}_{\parallel}=\gamma(\gamma(\vec{r}_{\parallel}-\vec{V}t)+\vec{V}t')=\gamma^{2}\vec{r}_{\parallel}-\gamma^{2}\vec{V}t+\gamma\vec{V}t',\\
& \therefore (1-\gamma^{2})\vec{V}\cdot\vec{r}_{\parallel}+\gamma^{2}\vec{V}\cdot\vec{V}t=\gamma\vec{V}\cdot\vec{V}t', \\
& \therefore t'=\gamma\left(t-\frac{\gamma^{2}-1}{\gamma^{2}}\frac{\vec{V}\cdot\vec{r}}{V^{2}}\right), \\
& \therefore t=\gamma\left(t'+\frac{\gamma^{2}-1}{\gamma}\frac{\vec{V}\cdot\vec{r'}}{V^{2}}\right). \ \text{(por P1)}
\end{align}

Para __obter $\gamma$__, usamos o postulado P2 aplicado ao experimento do pulso esférico de luz:
\begin{equation}
\vec{r}\cdot\vec{r}=r_{\perp}^{2}+r_{\parallel}^{2}=c^{2}t^{2} \text{ e } \vec{r'}\cdot\vec{r'}=c^{2}t'^{2}. 
\end{equation}
Assim, usando $\vec{V}\cdot\vec{r}=Vr_{\parallel}$, teremos
\begin{align}
\vec{r'}\cdot\vec{r'} & = (\vec{r'}_{\perp}+\vec{r'}_{\parallel})\cdot(\vec{r'}_{\perp}+\vec{r'}_{\parallel})  = [\vec{r}_{\perp}+\gamma(\vec{r}_{\parallel}-\vec{V}t)]\cdot[\vec{r}_{\perp}+\gamma(\vec{r}_{\parallel}-\vec{V}t)] \\ 
&= r_{\perp}^{2}+0+0+\gamma^{2}(r_{\parallel}^{2}-2\vec{r}_{\parallel}\cdot\vec{V}t+V^{2}t^{2}) \\
=c^{2}t'^{2}& = c^{2}\gamma^{2}\left(t-\frac{\gamma^{2}-1}{\gamma}\frac{\vec{V}\cdot\vec{r}}{V^{2}}\right)^{2}=c^{2}\gamma^{2}\left(t^{2}-2\frac{\gamma^{2}-1}{\gamma}\frac{r_{\parallel}}{V}t+\frac{(\gamma^{2}-1)^{2}}{\gamma^{2}}\frac{r_{\parallel}^{2}}{V^{2}}\right). \\
& \Rightarrow r_{\perp}^{2}+\left(\gamma^{2}-\frac{c^{2}(\gamma^{2}-1)^{2}}{V^{2}}\right)r_{\parallel}^{2} = \left(\gamma^{2}-\gamma^{2}V^{2}/c^{2}\right)c^{2}t^{2} + 2c^{2}\gamma^{2}\left(-\frac{\gamma^{2}-1}{\gamma V}+\frac{V}{c^{2}}\right)r_{\parallel}t, \\
& \Rightarrow r_{\perp}^{2}+Cr_{\parallel}^{2} = \left(\gamma^{2}-\gamma^{2}V^{2}/c^{2}\right)c^{2}t^{2} + Dr_{\parallel}t, \\
& \Rightarrow C=1,D=0 \text{ e } \gamma^{2}-\gamma^{2}V^{2}/c^{2}=1 \Rightarrow \gamma=\gamma(V)=\frac{1}{\sqrt{1-V^{2}/c^{2}}}.
\end{align}

### Boost de Lorentz 3D
Vamos escrever
\begin{equation}
\vec{\rho'}=\Lambda(\vec{V})\vec{\rho}\therefore\begin{bmatrix}\vec{r'}\\ ct'\end{bmatrix}=\begin{bmatrix}\Lambda_{11}&\Lambda_{12}\\\Lambda_{21}&\Lambda_{22}\end{bmatrix}\begin{bmatrix}\vec{r}\\ ct\end{bmatrix}.
\end{equation}
Vamos definir $\vec{\beta}:=\vec{V}/c$ e lembrar que em $\mathbb{R}^{3}$ o produto interno é $$\vec{x}\cdot\vec{y}:=\vec{x}^{T}\vec{y}.$$ Assim
\begin{align}
\vec{r'} & = \vec{r'}_{\perp} + \vec{r'}_{\parallel} = \vec{r}_{\perp}+\gamma(\vec{r}_{\parallel}-\gamma\vec{V}t) = \vec{r}-\vec{r}_{\parallel}+\gamma\vec{r}_{\parallel}-\gamma(\vec{V}/c)ct \\
& = \vec{r}+(\gamma-1)\vec{r}\cdot(\vec{V}/V)(\vec{V}/V)-\gamma(\vec{V}/c)ct = \vec{r}+(\gamma-1)\frac{\vec{V}/c}{V/c}\frac{\vec{V}/c}{V/c}\cdot\vec{r}-\gamma(\vec{V}/c)ct \\
& = \left(\mathbb{I}_{3}+(\gamma-1)\vec{\beta}\vec{\beta}^{T}/\beta^{2}\right)\vec{r}-\gamma\vec{\beta}ct.
\end{align}
e
\begin{align}
ct'=\gamma\left(ct-c\frac{\gamma^{2}-1}{\gamma}\frac{\vec{V}\cdot\vec{r}}{V^{2}}\right)=\gamma\left(ct-\frac{V^{2}}{c}\frac{\vec{V}\cdot\vec{r}}{V^{2}}\right) = -\gamma\vec{\beta}^{T}\vec{r} + \gamma ct.
\end{align}
O __boost de Lorentz 3D__ fica escrito na forma matricial como segue:
\begin{equation}
\Lambda(\vec{\beta})=\begin{bmatrix} \left(\mathbb{I}_{3}+(\gamma(\beta)-1)\vec{\beta}\vec{\beta}^{T}/\beta^{2}\right) & -\gamma(\beta)\vec{\beta} \\ -\gamma(\beta)\vec{\beta}^{T} & \gamma(\beta) \end{bmatrix},
\end{equation}
em que
\begin{equation}
\gamma(\beta)=\frac{1}{\sqrt{1-\beta^{2}}} \text{ e } \beta=\frac{V}{c}.
\end{equation}

### Composição relativística de velocidades no "caso 3D"
Nesse caso teremos
\begin{align}
& d\vec{r'}= \left(\mathbb{I}_{3}+(\gamma-1)\vec{\beta}\vec{\beta}^{T}/\beta^{2}\right)d\vec{r}-\gamma\vec{\beta}cdt, \\
& cdt' = -\gamma\vec{\beta}^{T}d\vec{r} + \gamma cdt.
\end{align}
Assim
\begin{align}
\vec{u'} & = \frac{d\vec{r'}}{dt'} = \frac{\left(\mathbb{I}_{3}+(\gamma-1)\vec{\beta}\vec{\beta}^{T}/\beta^{2}\right)d\vec{r}-\gamma\vec{\beta}cdt}{-\gamma\vec{\beta}^{T}d\vec{r}/c + \gamma dt} \\
&= \frac{\left(\mathbb{I}_{3}+(\gamma-1)\vec{\beta}\vec{\beta}^{T}/\beta^{2}\right)\vec{u}-\gamma\vec{\beta}c}{\gamma (1-\vec{\beta}^{T}\vec{u}/c)}  = \frac{\left(\mathbb{I}_{3}+(\gamma-1)\hat{V}\hat{V}^{T}\right)\vec{u}-\gamma\vec{V}}{\gamma (1-\vec{V}^{T}\vec{u}/c^{2})} \\ 
&= \frac{\vec{u} + \left((\gamma-1)\hat{V}^{T}\vec{u}/V -\gamma\right)\vec{V}}{\gamma \left(1-\vec{V}^{T}\vec{u}/c^{2}\right)}.
\end{align}

Podemos reescrever essa última equação como:
\begin{align}
& \gamma\vec{u'}-\gamma\vec{u'}\vec{V}^{T}\vec{u}/c^{2}=\left(\mathbb{I}_{3}+(\gamma-1)\hat{V}\hat{V}^{T}\right)\vec{u}-\gamma\vec{V}, \\
& \therefore\gamma(\vec{u'}+\vec{V})=\left(\mathbb{I}_{3}+(\gamma-1)\hat{V}\hat{V}^{T} +\gamma\vec{u'}\vec{V}^{T}/c^{2}\right)\vec{u}.
\end{align}

### Composição de boosts de Lorentz 3D
Vamos considerar um referencial inercial $O''$ se movimentando com "velocidade" $\vec{\alpha}:=\vec{V'}/c$ em relação ao RI $O'$, que por sua vez se movimenta com "velocidade" $\vec{\beta}:=\vec{V}/c$ em relação ao referencial inercial $O$. Teremos a seguinte relação entre os vetores no espaço-tempo:
\begin{equation}
\rho''=\Lambda(\vec{\alpha})\rho'=\Lambda(\vec{\alpha})\Lambda(\vec{\beta})\rho.
\end{equation}
Se a composição de dois boosts de Lorentz for um boost de Lorentz, teremos que
\begin{equation}
\rho''=\Lambda(\vec{\eta})\rho \text{ com } \Lambda(\vec{\eta})=\Lambda(\vec{\alpha})\Lambda(\vec{\beta})
\end{equation}
para $\vec{\eta}:=\vec{V''}/c$ a "velocidade" relativa entre os RIs $O''$ e $O$. Segue que
\begin{align}
\Lambda(\vec{\alpha})\Lambda(\vec{\beta}) & \begin{bmatrix} \left(\mathbb{I}_{3}+(\gamma(\alpha)-1)\vec{\alpha}\vec{\alpha}^{T}/\alpha^{2}\right) & -\gamma(\alpha)\vec{\alpha} \\ -\gamma(\alpha)\vec{\alpha}^{T} & \gamma(\alpha) \end{bmatrix}\begin{bmatrix} \left(\mathbb{I}_{3}+(\gamma(\beta)-1)\vec{\beta}\vec{\beta}^{T}/\beta^{2}\right) & -\gamma(\beta)\vec{\beta} \\ -\gamma(\beta)\vec{\beta}^{T} & \gamma(\beta) \end{bmatrix} \\
&= \begin{bmatrix} \left(\left(\mathbb{I}_{3}+(\gamma(\alpha)-1)\vec{\alpha}\vec{\alpha}^{T}/\alpha^{2}\right)\left(\mathbb{I}_{3}+(\gamma(\beta)-1)\vec{\beta}\vec{\beta}^{T}/\beta^{2}\right)+\gamma(\alpha)\gamma(\beta)\vec{\alpha}\vec{\beta}^{T}\right)
& -\gamma(\beta)\left(\vec{\beta}+\left(\gamma(\alpha)+(\gamma(\alpha)-1)\vec{\alpha}^{T}\vec{\beta}/\alpha^{2}\right)\vec{\alpha}\right) \\
-\gamma(\alpha)\left(\vec{\alpha}^{T}+\left(\gamma(\beta)+(\gamma(\beta)-1)\vec{\alpha}^{T}\vec{\beta}/\beta^{2}\right)\vec{\beta}^{T}\right) 
& \gamma(\alpha)\gamma(\beta)(1+\vec{\alpha}^{T}\vec{\beta}) \end{bmatrix}.
\end{align}
Usando a decomposição relatística de velocidades no caso geral, tratada acima, podemos verificar que
\begin{equation}
\vec{\eta} = \gamma(\beta)(\vec{\alpha}+\vec{\beta})\left(\mathbb{I}_{3}+(\gamma(\beta)-1)\left(\vec{\beta}/\beta^{2}+\vec{\alpha}\right)\vec{\beta}^{T}\right)^{-1}.
\end{equation}
O elemento de matriz "mais fácil" de ser verificado seria
\begin{equation}
\gamma(\vec{\eta})=\gamma(\alpha)\gamma(\beta)(1+\vec{\alpha}^{T}\vec{\beta})?
\end{equation}
No entanto, a forma complicada de $\vec{\eta}$ indica que essa tarefa não é factível. Na realidade, sabe-se que para o caso geral, boosts de Lorentz precisam ser combinados com rotações para termos estrutura de grupo.
